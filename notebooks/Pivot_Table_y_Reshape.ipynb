# Pivot Table y Reshape

```python
import pandas as pd

# Leer el dataset de ventas desde un archivo CSV
file_path =  './content/online_retail.csv'
sales_data = pd.read_csv(file_path)

pivot_table = pd.pivot_table(sales_data, values = 'Quantity', index = 'Country',
                             columns='StockCode', aggfunc='sum')
print(pivot_table)
```

```python
df = pd.DataFrame({
    'A': ['foo', 'bar', 'baz'],
    'B': [1, 2, 3],
    'C': [4, 5, 6]
})
print(df)
```

```python
df_stacked = df.stack()
print(df_stacked)
```

```python
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [4, 5, 6]
})
print(df1)
print(df2)
```

```python
inner_merged = pd.merge(df1, df2, on='key', how='inner')
print(inner_merged)
```

```python
outer_merged = pd.merge(df1, df2, on='key', how='outer')
print(outer_merged)
```

```python
left_merged = pd.merge(df1, df2, on='key', how='left')
print(left_merged)
```

```python
right_merged = pd.merge(df1, df2, on='key', how='right')
print(right_merged)
```

```python
df3 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
})
df4 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5']
})
print(df3)
print(df4)
```

```python
vertical_concat = pd.concat([df3, df4])
print(vertical_concat)
```

```python
horizontal_concat = pd.concat([df3, df4], axis=1)
print(horizontal_concat)
```

```python
df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
}, index=['K0', 'K1', 'K2'])
df6 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])
print(df5)
print(df6)
```

```python
joined = df5.join(df6, how='inner')
print(joined)
```
