# Series temporales

```python
import pandas as pd

file_path =  './content/online_retail.csv'

df = pd.read_csv(file_path)


print("Primeras filas del dataset original")
df.head()

#Convertir la columna a tipo datetime
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df.dropna(subset=['InvoiceDate'], inplace= True)
```

```python
df.set_index('InvoiceDate', inplace= True)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Weekday'] = df.index.weekday
df['Hour'] = df.index.week

print(df['Year'])
print(df['Month'])
print(df['Day'])
print(df['Weekday'])
print(df['Hour'])
```

```python
df_2011_dec = df.loc['2011-12']
df_2011_dec.head()
```

```python
df_dec_range = df.loc['2010-12-01':'2010-12-15']
df_dec_range.head()
```

```python
date_range_new = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
date_range_new
```

```python
df_dates = pd.DataFrame(date_range_new, columns=['Date'])
df_dates.head()
```

```python
monthly_sales = df['Quantity'].resample('M').sum()
print(monthly_sales)
```

```python
max_sales_month = monthly_sales.idxmax()
max_sales_value = monthly_sales.max()

print(f"El mes con mayores ventas es {max_sales_month.strftime('%B %Y')} con un total de {max_sales_value} unidades vendidas.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Cantidad Vendida')
plt.show()
```
